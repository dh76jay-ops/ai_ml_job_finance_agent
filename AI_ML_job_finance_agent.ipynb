```python
from smolagents import CodeAgent, InferenceClientModel, DuckDuckGoSearchTool, tool
import os
token = os.getenv("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")
model = InferenceClientModel(
    model_id="Qwen/Qwen2.5-72B-Instruct",
    token=token
)

print("Model ready!")
```

```python
@tool
def ai_ml_salary_insights(role: str, experience_level: str) -> str:
    """
    AI/ML job roles ke liye salary insights deta hai.
    
    Args:
        role: Job role jaise 'ML Engineer', 'Data Scientist'
        experience_level: 'fresher', 'mid', 'senior'
    """
    query = f"{role} {experience_level} salary India 2024 AI ML"
    search_tool = DuckDuckGoSearchTool()
    results = search_tool(query)
    return results

print("Tool ready!")
```

```python
tools = [DuckDuckGoSearchTool(), ai_ml_salary_insights]

agent = CodeAgent(
    tools=tools,
    model=model,
    max_steps=7  # 5 se badha ke 7 kiya - complex queries ke liye
)

print("Agent ready!")
```

```python
# "Capital of France" nahi - apna real sawaal
result = agent.run(
    "What skills should a fresher AI/ML engineer learn in 2024 to get hired at top tech companies? Also what is the average salary range in India?"
)
print(result)
```
